```python
# Password Generate Python Project Module

import random

alpha = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
number = "0123456789"
special = "!@#$%^&*()_+-=[]{}|;:,.<>/?"

print("***Welcome to the Password Generator***")

user_l = input("Include letters? (y/n): ")
user_n = input("Include numbers? (y/n): ")
user_s = input("Include symbols? (y/n): ")

character = ""

if user_l.lower() == 'y':
  character += alpha
if user_n.lower() == 'y':
  character += number
if user_s.lower() == 'y':
  character += special

if character == '':
  print("Please select at least one character type.")
  exit()

count = int(input("How many password do you want to generate? "))
length = int(input("Enter the desired length of your password: "))
print()

for j in range(count):
  password = ""

  for i in range(length):
    password += random.choice(character)

  if length <= 6:
    strength = 'weak'
  elif length <=10:
    strength = 'medium'
  else:
    strength = 'strong'

  print("Password", j+1, ":", password, "| Strength:", strength)

print("\nThank you for using the Password Generator!")

```
