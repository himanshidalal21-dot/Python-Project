```python
import random

letters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
numbers = "0123456789"
symbols = "!@#$%^&*()_+-=[]{}|;:,.<>/?"

print("Welcome to the Advanced Password Generator!")

# User choices
use_letters = input("Include letters? (y/n): ")
use_numbers = input("Include numbers? (y/n): ")
use_symbols = input("Include symbols? (y/n): ")

# Create character pool
characters = ""

if use_letters.lower() == "y":
    characters += letters
if use_numbers.lower() == "y":
    characters += numbers
if use_symbols.lower() == "y":
    characters += symbols
print()

if characters == "":
    print("You must select at least one character type.")

else:
    count = int(input("How many passwords do you want to generate? "))
    length = int(input("Enter the desired password length: "))

    for j in range(count):
        password = ""

        for i in range(length):
            password += random.choice(characters)

        # Password strength check
        if length < 6:
            strength = "Weak"
        elif length <= 10:
            strength = "Medium"
        else:
            strength = "Strong"

        print("Password", j+1, ":", password, "| Strength:", strength)

print("\nThank you for using the Password Generator!")
```
